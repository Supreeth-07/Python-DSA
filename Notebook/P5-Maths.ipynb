### Sum of N Even Natural Numbers
n(n+1)

```python
def sum_of_n_even_numbers(n):
    return n * (n + 1)

print(sum_of_n_even_numbers(5))    
```

### Check for Even Number

```python
def is_even(num):
    return num % 2 == 0

print(is_even(4))
print(is_even(7))
```

### Check for Prime Number

```python
def is_prime(num):
    
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

print(is_prime(7))
print(is_prime(12))
```

### Valid Perfect Square

```python
import math

def is_perfect_square(num):
    if num < 0:
        return False
    root = int(math.sqrt(num))
    return root * root == num

print(is_perfect_square(25))
print(is_perfect_square(26))    
```

### Decimal to Binary


```python
def decimal_to_binary(n):
    return bin(n)[2:]

print(decimal_to_binary(10))    
```

### Binary to Decimal

```python
def binary_to_decimal(b):
    return int(b, 2)

print(binary_to_decimal("1010"))
```
