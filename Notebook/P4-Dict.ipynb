# Dictionaries

### Merge Multiple Dictionaries
Example:

Input: ({'a': 1, 'b': 2}, {'c': 3, 'd': 4}, {'e': 5, 'f': 6})
Output: {'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5, 'f': 6}

Input: ({'x': 10, 'y': 20}, {'z': 30}, {'a': 40, 'b': 50})
Output: {'x': 10, 'y': 20, 'z': 30, 'a': 40, 'b': 50}

```python
def merge_dicts(*dicts):
    merged = {}

    for d in dicts:
        merged.update(d)  

    return merged


result1 = merge_dicts({'a': 1, 'b': 2}, {'c': 3, 'd': 4}, {'e': 5, 'f': 6})
print(result1)

result2 = merge_dicts({'x': 10, 'y': 20}, {'z': 30}, {'a': 40, 'b': 50})
print(result2)
```

### Words Frequency in a Sentence
Count Word Frequency

Input: "hello world hello"
Output: {'hello': 2, 'world': 1}

```python
def word_count(text):
    words = text.split()  # Split the string into words
    count = {}
    for word in words:
        count[word] = count.get(word, 0) + 1  # Increment word count
    return count


result1 = word_count("hello world hello")
print(result1)

result2 = word_count("the quick brown fox jumps over the lazy dog")
print(result2)
```
