### Handle missing data
Use fillna(), dropna(), and isna()

```python
df = pd.DataFrame({'A': [1, None, 3], 'B': [4, 5, None]})

print(df)
print('-'*25)

df1 = df.fillna(0)
print(df1)
print('-'*25)

df2 = df.dropna()
print(df2)
print('-'*25)

df3 = df.isna().sum()
print(df3)
```

| Method/Option                         | Description                                      | Example                                                        |
|--------------------------------------|--------------------------------------------------|----------------------------------------------------------------|
| `isna()` / `notna()`                  | Detect missing values                             | `df.isna()`                                                    |
| `fillna(value=..., method=..., limit=...)` | Fill NaN with value/statistics/forward/backward | `df.fillna({'Age': df['Age'].mean(), 'Name':'Unknown'})`       |
| `interpolate(method=...)`             | Fill numeric NaN using interpolation             | `df['Age'].interpolate()`                                      |
| `dropna(axis=0/1, how=any/all, subset=..., thresh=...)` | Drop rows/columns with missing values | `df.dropna(subset=['Age'], how='any')`                         |
| Flagging missing                      | Create indicator for missing values              | `df['Age_missing'] = df['Age'].isna()`                         |


### Filter and select data

| Operation                     | Example                                     | Description                                  |
|-------------------------------|---------------------------------------------|----------------------------------------------|
| Select column                 | `df['Name']`                                | Get a single column                           |
| Select multiple columns       | `df[['Name','City']]`                       | Get multiple columns                          |
| Select row by index           | `df.iloc[0]`                                | Get row by position                           |
| Select row by label           | `df.loc[1]`                                 | Get row by index label                        |
| Conditional filter            | `df[df['Age'] > 28]`                        | Filter rows by condition                      |
| Multiple conditions           | `df[(df['Age']>28) & (df['City']=='SF')]`  | Filter using AND / OR                         |
| Filter with `isin()`          | `df[df['City'].isin(['NY','LA'])]`          | Filter for multiple possible values          |
| Filter with string methods    | `df[df['Name'].str.startswith('A')]`        | Filter based on string pattern               |
| Rows & columns together       | `df.loc[df['Age']>28, ['Name','City']]`     | Select specific rows and columns at once     |


```python
df = pd.DataFrame({'Name': ['Alice', 'Bob'], 'Age': [25, 30]})
print(df['Name'])
print("-"*25)
print(df[df['Age'] > 25])
```

### Group data and get summary
Use groupby() for aggregation.

```python
df = pd.DataFrame({'Team': ['A', 'A', 'B'], 'Score': [10, 20, 30]})
print(df.groupby('Team')['Score'].mean())
```

### DataFrame operations

| Operation                  | One-line Description                                   | Example |
|-----------------------------|--------------------------------------------------------|---------|
| Build DataFrame             | Create a new DataFrame with defined columns           | `df = pd.DataFrame(columns=['Name','Age','City'])` |
| Generate DataFrame          | Read CSV/Excel into a DataFrame                        | `df = pd.read_csv('data.csv')` |
| Add Data Column             | Add a new column to an existing DataFrame             | `df['Salary'] = [50000,60000,55000]` |
| Add Data Row                | Insert a new row into a DataFrame                      | `df.loc[len(df)] = ['Alice',25,'NY']` |
| For Each Row                | Loop through each row for processing                  | `for idx, row in df.iterrows(): print(row['Name'])` |
| Get Row Item                | Get the value of a specific column in a row          | `age = df.loc[0, 'Age']` |
| Assign                      | Update values in a specific row or column            | `df.loc[0, 'Age'] = 26` |
| Filter DataFrame            | Extract rows that meet a condition                    | `df[df['Age']>30]` |
| Select                      | Get rows based on a filter expression                 | `df.loc[df['City']=='NY']` |
| Sort Data Table             | Sort rows by one or more columns                      | `df.sort_values(by='Age', ascending=False)` |
| Compute                     | Perform aggregate calculations on a column           | `df['Salary'].mean()` |
| Lookup Data Table           | Find a value using a key                               | `df.loc[df['Name']=='Alice', 'Salary'].values[0]` |
| Remove Data Row              | Delete a specific row                                  | `df.drop(index=0, inplace=True)` |
| Clear Data Table             | Remove all rows from the DataFrame                     | `df.drop(df.index, inplace=True)` |
| Merge DataFrame             | Combine two DataFrames into one                        | `pd.concat([df1, df2], ignore_index=True)` |
| Join Data Tables            | Perform inner/left/right join between DataFrames      | `df1.merge(df2, on='ID', how='left')` |
| Clone DataFrame             | Copy only structure (columns) without data           | `df_clone = pd.DataFrame(columns=df.columns)` |
| Copy DataFrame              | Copy structure and all data                            | `df_copy = df.copy()` |
| Remove Duplicate Rows       | Remove repeated entries                                 | `df.drop_duplicates(inplace=True)` |
| Output Data Table           | Convert DataFrame to string for display                | `print(df.to_string())` |
| Write Range                 | Export DataFrame to CSV or Excel                        | `df.to_csv('output.csv', index=False)` |

### How to apply a function on a column?

Use apply() or vectorized operations.

```python
df = pd.DataFrame({'Salary': [50000, 60000]})

df1 = df

df1['Tax'] = df1['Salary'].apply(lambda x: x * 0.2)

print(df1)
print('-'*25)

##############################

df2 = df

def calculate_tax(salary):
    return salary * 0.2

df2['Tax'] = df2['Salary'].apply(calculate_tax)

print(df2)
print('-'*25)

##############################

df3 = df

df3['Tax'] = df3['Salary'] * 0.2

print(df3)
```

### Sort and rank data
Use sort_values() and rank() methods.

```python
df = pd.DataFrame({'Name': ['Bob', 'Alice'], 'Score': [80, 90]})
print(df)
print('-'*22)

df1 = df.sort_values(by='Score', ascending=False)
print(df1)
print('-'*22)

df2 = df
df2['Rank'] = df2['Score'].rank(ascending=False)
print(df2)
```

### Pivot and reshape data
Use pivot() or melt() for reshaping tables.

```python
df = pd.DataFrame({
    'Month': ['Jan', 'Jan', 'Feb', 'Feb'],
    'City': ['A', 'B', 'A', 'B'],
    'Sales': [100, 150, 200, 250]
})

print(df)
print('-'*22)

df.pivot(index='Month', columns='City', values='Sales')
```

### Correlation between columns?

Use .corr() to measure relationships between numeric columns.

```python
df = pd.DataFrame({'A': [1, 2, 3], 'B': [2, 4, 6]})
print(df.corr())
```

### Get top N records from each group?

Use groupby() + nlargest().

```python
df = pd.DataFrame({
    'Month': ['Jan', 'Jan', 'Feb', 'Feb'],
    'City': ['A', 'B', 'A', 'B'],
    'Sales': [100, 150, 200, 250]
})

df2 = df.groupby('Month')['Sales'].nlargest(2)

print(df2)
```
