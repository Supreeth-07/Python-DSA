# String

### Reverse a String

```python
def reverse_string(s: str) -> str:
    return s[::-1]

print(reverse_string("hello")) 
```

### Count Vowels in a String

```python
def count_vowels(s: str) -> int:
    vowels = "aeiouAEIOU"
    return sum(1 for char in s if char in vowels)

print(count_vowels("Hello World"))
```

### Check for Same Strings

```python
def are_same_strings(s1: str, s2: str) -> bool:
    if s1 == s2:
        return True
    else: 
        return False


print(are_same_strings("abc", "123"))
print(are_same_strings("abc", "abc"))
```

### Check Palindrome

```python
def is_palindrome(s: str) -> bool:
    s = s.lower().replace(" ", "")
    return s == s[::-1]

print(is_palindrome("Racecar"))
print(is_palindrome("Racecars"))
```

### Count Words in a String

```python
def count_words(s: str) -> int:
    return len(s.split())

print(count_words("Count the words in this sentence"))  # 6
```

### Remove Duplicates in a String

```python
def remove_duplicates(s: str) -> str:
    result = ""
    for char in s:
        if char not in result:
            result += char
    return result

print(remove_duplicates("banana"))
```

### Count Consonants in a String

```python
def count_consonants(s):
    vowels = "aeiouAEIOU"
    count = 0
    for ch in s:
        if ch.isalpha() and ch not in vowels:
            count += 1
    return count


print(count_consonants("Hello World"))  
```

### Check for Anagrams


Given two strings, check whether they are anagrams (contain the same letters with the same frequency).

```python
def are_anagrams(s1, s2):
    return sorted(s1.replace(" ", "").lower()) == sorted(s2.replace(" ", "").lower())


print(are_anagrams("listen", "silent"))  # Output: True
```

### Check for Substring

```python
def is_substring(s, sub):
    return sub in s


print(is_substring("hello", "ell"))
```

### Length of the Longest Word

```python
def longest_word_length(sentence):
    words = sentence.split()
    return max(len(word) for word in words) if words else 0

# Exampl
print(longest_word_length("The quick brown fox"))  
```
